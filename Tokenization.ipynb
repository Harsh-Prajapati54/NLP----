```python
device = "cuda"
```

```python
from transformers import AutoTokenizer , AutoModelForCausalLM

model = AutoModelForCausalLM.from_pretrained(
    "microsoft/phi-3-mini-4k-instruct",
    torch_dtype = "auto",
    device_map = "cuda",
    trust_remote_code = False,
)


Tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-3-mini-4k-instruct")
```

```python
prompt = "write an email to Tushar for inviting a 5 day trip to  Varanashi "

# tokenize the input prompt
input_ids = Tokenizer(prompt,return_tensors="pt").input_ids.to(device)

# generate the text
generation_output = model.generate(
    input_ids=input_ids,
    max_new_tokens=256,

)

# print the output
print(Tokenizer.decode(generation_output[0]))
```

## Compairing Diffrent Tokenizer

we have `text` which contain the diffrent type of character like , Capital , Emojies and codes .

### Types of  Special Tokens
unk_token = [UNK]  

* an unknown token that tokenizer dont
have encoding

sep_token  = [SEP]

* A seperator that enables certain tasks that require giving the model two text

pad_token = [PAD]

* a padding token used to pad unused position in the model's input  

cls_token = [CLS]

* A special classification token for classification

mask_token = [mask]

* a masking token is used to hide the token during training

```python
text = """
          English and CAPITILIZATION

          😒😂🤣

          shown_tokens False None elif == >=  else :
          two tabs: ""  Three tabs:""

          20122005

          """
```

```python
from transformers import AutoTokenizer , AutoModelForCausalLM

color_list = [
    '102;194;165', '252;141;98', '141;160;203',
    '231;138;195', '166;216;84', '255;217;47'
]

def  show_tokens(sentence,tokenizer_name):
  Tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
  token_ids = Tokenizer(sentence).input_ids

  for idx, t in enumerate(token_ids):
    print(
        f'\x1b[0;30;48;2;{color_list[idx % len(color_list)]}m' +
            Tokenizer.decode(t) +
            '\x1b[0m',
            end=' '
    )
```

Bert base Uncased Tokenization

in this type of tokenization the capital letter is gone or converted into smaller case

`##` means the token is connected to preceding token

Voculabary_size = 30,522

```python
show_tokens(text,"bert-base-uncased")
```

```python
show_tokens(text,"bert-base-cased")
```

in this tokenization the capital letter is not converted into lowercase letter


vocubulary size = 28,996

```python
show_tokens(text,"gpt2")
```
